```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# 1. Download data
def download_data(tickers, period="6mo", interval="1d"):
    data = {}
    for ticker in tickers:
        try:
            df = yf.download(ticker, period=period, interval=interval)
            df.dropna(inplace=True)
            data[ticker] = df
        except Exception as e:
            print(f"Error downloading {ticker}: {e}")
    return data

# 2. Apply Strategy (Add indicators)
def apply_strategy(df):
    df = df.copy()
    df["SMA20"] = df["Close"].rolling(window=20).mean()
    df["SMA50"] = df["Close"].rolling(window=50).mean()
    return df

# 3. Generate Buy/Sell Signals
def generate_signals(df):
    df = df.copy()
    df["Signal"] = 0
    for i in range(1, len(df)):
        if df["SMA20"].iloc[i] > df["SMA50"].iloc[i] and df["SMA20"].iloc[i - 1] <= df["SMA50"].iloc[i - 1]:
            df.loc[df.index[i], "Signal"] = 1  # Buy
        elif df["SMA20"].iloc[i] < df["SMA50"].iloc[i] and df["SMA20"].iloc[i - 1] >= df["SMA50"].iloc[i - 1]:
            df.loc[df.index[i], "Signal"] = -1  # Sell
    return df

# 4. Plotting
def plot_closing_prices(df, ticker):
    plt.figure(figsize=(14, 6))
    plt.plot(df.index, df["Close"], label="Closing Price", color="blue")
    plt.plot(df.index, df["SMA20"], label="SMA 20", color="green", linestyle="--")
    plt.plot(df.index, df["SMA50"], label="SMA 50", color="red", linestyle="--")

    # Buy signals
    plt.plot(df[df["Signal"] == 1].index, df[df["Signal"] == 1]["Close"], '^', markersize=10, color='g', label='Buy Signal')
    # Sell signals
    plt.plot(df[df["Signal"] == -1].index, df[df["Signal"] == -1]["Close"], 'v', markersize=10, color='r', label='Sell Signal')

    plt.title(f"{ticker} Price Chart with SMA and Buy/Sell Signals")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 5. Run the full pipeline
tickers = ["RELIANCE.NS"]  # You can add more tickers here
stock_data = download_data(tickers)

for ticker in tickers:
    if ticker in stock_data:
        df = stock_data[ticker]
        df = apply_strategy(df)
        df = generate_signals(df)
        plot_closing_prices(df, ticker)
    else:
        print(f"No data for {ticker}")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def fetch_stock_data(symbol, start, end):
    df = yf.download(symbol, start=start, end=end)
    df['SMA20'] = df['Close'].rolling(window=20).mean()
    df['SMA50'] = df['Close'].rolling(window=50).mean()
    return df

def generate_signals(df):
    signals = []
    position = False  # Not holding any stock initially

    for i in range(1, len(df)):
        # Use iloc for index-safe access
        if df['SMA20'].iloc[i] > df['SMA50'].iloc[i] and df['SMA20'].iloc[i - 1] <= df['SMA50'].iloc[i - 1] and not position:
            signals.append((df.index[i], df['Close'].iloc[i], 'Buy'))
            position = True
        elif df['SMA20'].iloc[i] < df['SMA50'].iloc[i] and df['SMA20'].iloc[i - 1] >= df['SMA50'].iloc[i - 1] and position:
            signals.append((df.index[i], df['Close'].iloc[i], 'Sell'))
            position = False

    return signals

def plot_signals(df, signals):
    plt.figure(figsize=(14, 7))
    plt.plot(df['Close'], label='Close Price', alpha=0.5)
    plt.plot(df['SMA20'], label='SMA20', alpha=0.9)
    plt.plot(df['SMA50'], label='SMA50', alpha=0.9)

    for date, price, signal_type in signals:
        if signal_type == 'Buy':
            plt.scatter(date, price, label='Buy Signal', marker='^', color='green')
        elif signal_type == 'Sell':
            plt.scatter(date, price, label='Sell Signal', marker='v', color='red')

    plt.title('Buy/Sell Signals using SMA Crossover')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
symbol = 'AAPL'
start = '2022-01-01'
end = '2023-01-01'

df = fetch_stock_data(symbol, start, end)
signals = generate_signals(df)
plot_signals(df, signals)
```

```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to fetch stock data using yfinance
def fetch_stock_data(symbol, start_date='2022-01-01', end_date='2023-01-01'):
    """
    Fetch historical stock data for a given symbol using Yahoo Finance.
    """
    df = yf.download(symbol, start=start_date, end=end_date)
    df['Symbol'] = symbol
    return df

# Function to generate buy and sell signals using SMA crossover strategy
def generate_signals(df):
    """
    Add SMA20 and SMA50 columns, then generate buy/sell signals based on crossover.
    """
    df['SMA20'] = df['Close'].rolling(window=20).mean()
    df['SMA50'] = df['Close'].rolling(window=50).mean()

    # Signal: 1 = Buy, -1 = Sell
    df['Signal'] = 0
    df['Signal'] = np.where(df['SMA20'] > df['SMA50'], 1, -1)

    # Position helps identify the actual change (Buy/Sell trigger)
    df['Position'] = df['Signal'].diff()

    return df

# Function to plot buy/sell signals on the price chart
def plot_signals(df, symbol):
    """
    Plot the closing price along with SMA lines and buy/sell signals.
    """
    plt.figure(figsize=(14, 7))
    plt.plot(df['Close'], label='Close Price', alpha=0.5)
    plt.plot(df['SMA20'], label='SMA 20', linestyle='--')
    plt.plot(df['SMA50'], label='SMA 50', linestyle='--')

    # Plot Buy signals
    buy_signals = df[df['Position'] == 2]
    plt.plot(buy_signals.index, buy_signals['Close'], '^', markersize=10, color='green', label='Buy Signal')

    # Plot Sell signals
    sell_signals = df[df['Position'] == -2]
    plt.plot(sell_signals.index, sell_signals['Close'], 'v', markersize=10, color='red', label='Sell Signal')

    plt.title(f"{symbol} - SMA Crossover Strategy")
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the complete workflow
symbol = 'AAPL'  # You can change this to any valid stock symbol
df = fetch_stock_data(symbol)
df = generate_signals(df)
plot_signals(df, symbol)
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Fetch stock data
def fetch_stock_data(symbol, start_date='2022-01-01', end_date='2023-01-01'):
    df = yf.download(symbol, start=start_date, end=end_date)
    df['Symbol'] = symbol
    return df

# 2. Generate SMA signals
def generate_signals(df):
    df['SMA20'] = df['Close'].rolling(window=20).mean()
    df['SMA50'] = df['Close'].rolling(window=50).mean()

    df['Signal'] = 0
    df['Signal'] = np.where(df['SMA20'] > df['SMA50'], 1, -1)
    df['Position'] = df['Signal'].diff()

    return df

# 3. Calculate Profit/Loss from trades
def calculate_profit_loss(df):
    trades = []
    buy_price = None
    buy_date = None

    for index, row in df.iterrows():
        # Access the scalar value from the 'Position' Series
        position_value = row['Position'].item() if isinstance(row['Position'], pd.Series) else row['Position']

        # Buy signal
        if position_value == 2:
            buy_price = row['Close'].item() if isinstance(row['Close'], pd.Series) else row['Close']
            buy_date = index

        # Sell signal
        elif position_value == -2 and buy_price is not None:
            sell_price = row['Close'].item() if isinstance(row['Close'], pd.Series) else row['Close']
            sell_date = index
            profit = sell_price - buy_price
            trades.append({
                'Buy Date': buy_date.date(),
                'Buy Price': round(buy_price, 2),
                'Sell Date': sell_date.date(),
                'Sell Price': round(sell_price, 2),
                'Profit/Loss': round(profit, 2)
            })
            buy_price = None  # reset

    trades_df = pd.DataFrame(trades)
    return trades_df

# 4. Plot chart with signals
def plot_signals(df, symbol):
    plt.figure(figsize=(14, 7))
    plt.plot(df['Close'], label='Close Price', alpha=0.5)
    plt.plot(df['SMA20'], label='SMA 20', linestyle='--')
    plt.plot(df['SMA50'], label='SMA 50', linestyle='--')

    buy_signals = df[df['Position'] == 2]
    sell_signals = df[df['Position'] == -2]

    plt.plot(buy_signals.index, buy_signals['Close'], '^', markersize=10, color='green', label='Buy Signal')
    plt.plot(sell_signals.index, sell_signals['Close'], 'v', markersize=10, color='red', label='Sell Signal')

    plt.title(f"{symbol} - SMA Crossover Strategy")
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# 5. Main
symbol = 'AAPL'
df = fetch_stock_data(symbol)
df = generate_signals(df)
trades_df = calculate_profit_loss(df)
plot_signals(df, symbol)

# 6. Show trade summary
print("\nTrade Summary (Buy  Sell Pairs with Profit/Loss):")
print(trades_df)

# 7. Show total profit
total_profit = trades_df['Profit/Loss'].sum()
print(f"\nTotal Profit/Loss:  {round(total_profit, 2)}")
```

```python
import yfinance as yf
import pandas as pd
import ta

# Download data
df = yf.download("AAPL", start="2024-01-01", end="2024-06-30")

# Ensure 'Close' is a 1D Series
close_series = df['Close'].squeeze()

# Compute indicators safely
df['RSI'] = ta.momentum.RSIIndicator(close_series).rsi()
df['MACD'] = ta.trend.MACD(close_series).macd()
df['Signal'] = ta.trend.MACD(close_series).macd_signal()

# Fill missing values
df = df.dropna()

# Create target: 1 if next day close > today, else 0
df['Target'] = (df['Close'].shift(-1) > df['Close']).astype(int)

# Features and labels
features = df[['RSI', 'MACD', 'Signal', 'Volume']]
labels = df['Target']

# Train/Test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)

# Evaluate
accuracy = model.score(X_test, y_test)
print(f"Prediction accuracy: {accuracy:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_accuracy = rf_model.score(X_test, y_test)
print(f"Random Forest accuracy: {rf_accuracy:.2f}")
```

```python
import matplotlib.pyplot as plt

# Show feature importances
importances = rf_model.feature_importances_

# Convert column names to string
plt.figure(figsize=(10, 6))
plt.bar([str(col) for col in features.columns], importances)
plt.title("Feature Importance")
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
